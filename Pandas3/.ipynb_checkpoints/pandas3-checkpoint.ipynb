# Pandas 3: Grouping

### Lab Objective: 
Many data sets contain categorical values that naturally sort the data into groups. Analyzing and comparing such groups is an important part of data analysis. In this lab we explore pandas tools for grouping data and presenting tabular data more compactly, primarily through grouby and pivot tables. Details are found at http://www.acme.byu.edu/wp-content/uploads/2017/08/Pandas3.pdf.

## Problem 1

Examine the following data sets from pydataset and answer the corresponding questions. Use visualizations to support your conclusions.
- `"iris"`, measurements of various species of iris flowers.
    1. Which species is easiest to distinguish from the others? How?
    2. Given iris data without a species label, what strategies could you use to identify the flowerâ€™s species?

```python
iris = data("iris")
iris.plot()
```

```python
species = iris.groupby("Species")
species[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]].mean().plot(kind="barh",
                xerr=species.std(), title=r"Properties per species")
plt.xlabel("Length")
plt.ylabel("Species Classification")
```

Note that the most distinguishable feature is that setosa has the shortest petal length, based on this plot. We can also see this more clearly in the following plots for each species.

Strategies to easily identify species are that the flower that has the petals with the short length length and shortest width is the setosa. The flower with the linfes sepal length and petal length is the virginica.

- `"poisons"`, experimental results of three different poisons and four different treatments.
    1. In general, which poison is most deadly? Which treatment is most effective?
    2. If you were poisoned, how would you choose the treatment if you did not know which poison it was? What if you did know which poison it was?
(Hint: group the data by poison, then group each subset by treatment.)

```python
poisons = data("poisons")
poisons.boxplot(grid=False, column="time",by="poison")
plt.ylabel("Survival time in units of 10 hours")
```

```python
poisons = data("poisons")
poisons.boxplot(grid=False, column="time",by="treat")
plt.ylabel("Survival time in units of 10 hours")
```

Based on the first boxplot, I would say that poison 3 is the most deadly because the survival time is the lowest. Based on the second boxplot, I would say that people who took treatment B on averaged survived longest. Thus treatment B is the most effective.

```python
types = poisons.groupby("poison")
poison1 = types.get_group(1) 
poison2 = types.get_group(2) 
poison3 = types.get_group(3) 
```

```python
# plt.subplot(131)
poison1.boxplot(grid=False, column="time", by="treat")
plt.title("Poison 1")
plt.ylabel("Survival time in units of 10 hours")

poison2.boxplot(grid=False, column="time", by="treat")
plt.ylabel("Survival time in units of 10 hours")
plt.title("Poison 2")

poison3.boxplot(grid=False, column="time", by="treat")
plt.ylabel("Survival time in units of 10 hours")
plt.title("Poison 3")
```

For all types of poison, the treatment with the average longest survival time is treatment B, thus whithout knowing which poison I had, I would choose treatment B.

- `"diamonds"`, prices and characteristics of almost 54,000 round-cut diamonds.
    1. How does the color and cut of a diamond affect its price?
    2. Of the diamonds with color `"H"`, those with a `"Fair"` cut sell, on average, for a higher price than those with an `"Ideal"` (superior) cut. What other factors could explain this unintuitive statistic?

```python
diamonds = data("diamonds")
cut = diamonds.groupby("cut")
color = diamonds.groupby("color")
cut_color = diamonds.groupby(["cut", "color"])
diamonds.boxplot(grid=False, column="price", by="cut", showfliers=False)
```

The premium cut sell on average the most expensive diamond, while the ideal cut is on average the least expensive diamond.

```python
diamonds.boxplot(grid=False, column="price", by="color", showfliers=False)
```

Although, D is graded as the best color, it is one of the least expensive diamonds in regrads to color. The worse the color, the higher the price. 

```python
ax= plt.subplot(111)
cut_color.get_group(("Fair", "H")).plot(kind="kde", y="carat", ax=ax, label="fair cut")
cut_color.get_group(("Ideal", "H")).plot(kind="kde", y="carat", ax=ax, label="ideal cut")
plt.xlabel("Carat")
plt.ylabel("Density")
```

From this plot, we can see that there is a higher density of small diamonds, becaause a higher quality cut results in a small diamond, and vice versa.

## Problem 2

Suppose that someone claims that the city from which a passenger embarked had a strong influence on the passengerâ€™s survival rate. Investigate this claim.
1. Check the survival rates of the passengers based on where they embarked from (given in the `"Embarked"` column).
2. Create a pivot table to examine survival rates based on both place of embarkment and gender.
3. What do these tables suggest to you about the significance of where people embarked in influencing their survival rate? Examine the context of the problem, and explain what you think this really means.
4. Investigate the claim further with at least two more pivot tables, exploring other criteria (e.g., class, age, etc.). Carefully explain your conclusions.

```python
titanic = pd.read_csv("titanic.csv")
titanic.pivot_table(values="Survived", index=["Embarked"])
```

The survival rates based on where they embarked from seem to show that those who embarked from C correlated with ahigher survival rate.

```python
titanic.pivot_table(values="Survived", index=["Embarked", "Sex"])
```

The survival rates are higher for females than males from all embarkment locations, but this is especially high in location C.

```python
titanic.pivot_table(values="Survived", index=["Embarked", "Sex"], aggfunc="count")
```

Note that there were significantly more survivors from embarkment location S, even though the survival rate was lower in S than in C.

Conclusion: It seems that females were more likely to survive in general, and the highest survival rate for both men and women occurred at Embarkment C, however the highest number of survivors for both genders was at location S. Women and children probably survived at a higher rate, because they were given first priority for life jackets and life boats. 

```python
titanic.pivot_table(values="Survived", index=["Embarked", "Sex",  "Pclass"], aggfunc="count")
```

```python
titanic.pivot_table(values="Survived", index=["Embarked", "Pclass"], aggfunc="count")
```

I noticed that  more women survived if they were in Pclass 1 at location c, however more women survived from Pclass 3 in the other embarkment locations. In general, more survivors were from Pclass 3.

```python
titanic.pivot_table(values="Survived", index=["Embarked", "Sex", "Age"], aggfunc="count")
```

From this pivot table, we can see that a higher number of survivors from all embarkments were an average age of 29.

- `"swiss"`,standardized fertility measures and socio-economic indicators for French-speaking provinces of Switzerland at about 1888.
    1. What is the relationship in the data between fertility rates and infant mortality?
    2. How are provinces that are predominantly Catholic different from non-Catholic provinces, if at all?
    3. What factors in the data are the most important for predicting fertility?

```python
swiss = data("swiss")
```

Note that description on each variable was not provided, so the units on each cateogry are not well defined.

```python
plt.scatter(swiss["Fertility"], swiss["Infant.Mortality"])
plt.xlabel("Fertility (proportion of population)")
plt.ylabel("Infant Mortality (proportion of population)")
plt.title("Fertility vs. Infant Mortality")
```

```python
catholic = swiss.groupby("Catholic")
# catholic1 = catholic.loc[:50, :]
ax = plt.subplot(111)
catholic[["Fertility", "Agriculture", "Examination", "Education", "Infant.Mortality"]].mean().plot(kind="kde", title=r"Catholic Characteristics", ax=ax)
plt.xlabel("Density")
plt.ylabel("Frequency")
plt.title("")
```

It looks like provinces with lower amount of Catholics have a higher infant mortality rate, and areas with a higher frequency of Catholics have a higher fertility rate. It seems that Catholics tend to have higher fertility rates compared to the rest of the population.
