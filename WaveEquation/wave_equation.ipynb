```python
import numpy as np
from scipy import sparse
from matplotlib import pyplot as plt
from matplotlib import animation
# %matplotlib inline|
from numpy import linalg as la
plt.switch_backend('qt5agg')
```

## Problem 1

```python
#Problem 1
J=50
M=50
a,b = 0,0.5
x = np.linspace(0, 1, J+1)
t = np.linspace(a,b,M+1)
u = lambda x: np.sin(2.*np.pi*x)
f_double_prime = lambda x: -4.*(np.pi**2)*np.sin(2.*np.pi*x)

u_t = lambda t: 0 + 0*t
analytic_u = lambda x, t: np.sin(2.*np.pi*x)*np.cos(2.*np.pi*t)
s=1
delta_x = x[1]-x[0]
delta_t = t[1]-t[0]

lamb = s*(delta_t/delta_x)

A = np.zeros((J-1,J-1))
np.fill_diagonal(A, 2.*(1-(lamb**2)))
offset= (lamb**2)*np.ones(J-2)
A += np.diag(offset, 1)
A += np.diag(offset, -1)

U_m_0 = u(x[1:-1])
U_m_1 = U_m_0 + u_t(x[1:-1])+(s**2)*f_double_prime(x[1:-1])*((delta_t**2)/2.)


U_ans = np.zeros((J+1,M+1))
U_ans[1:-1,0] = U_m_0
U_ans[1:-1, 1] = U_m_1

for i in range(2, M+1):
    U_ans[1:-1,i] = np.dot(A, U_ans[1:-1, i-1]) - U_ans[1:-1,i-2]

# print (U_ans)
def initial_animation(res=51):
    y = U_ans
    f = plt.figure()
    plt.axes(xlim=(0,1), ylim=(-1,1))
    line, = plt.plot([], [])
    def animate(i):
        line.set_data(x, y[:,i])
        return line,

    a = animation.FuncAnimation(f, animate, frames=51, interval=20)
    plt.show()

initial_animation()
```
