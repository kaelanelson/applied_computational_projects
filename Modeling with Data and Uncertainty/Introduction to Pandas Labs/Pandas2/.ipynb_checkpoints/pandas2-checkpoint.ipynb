### 1) Trees Data Set 

This data set is formatted with 31 observations on 3 variables. It provides measurements of the girth, height and volume of timber in 31 felled black cherry trees. Note that girth is the diameter of the tree (in inches) measured at 4 ft 6 in above the ground.

```python
# Source: Ryan, T. A., Joiner, B. L. and Ryan, B. F. (1976) _The Minitab Student Handbook_. Duxbury Press. 
trees = data("trees")
```

We first create a simple histogram that visualizes the amount of trees with a given diameter measured in inches. It looks like the most common trees have diameters of 18 inches, and between 10-12 inches. One hypothesis for this occurence could be that the size od diameter may correlate with the location of where the trees grew. 

```python
trees[["Girth"]].hist(grid=False, bins = 50, alpha=.7)
plt.ylabel("Frequency of Trees")
plt.xlabel("Tree Diameter (inches)")
```

The scatter plot below displays a different relationship of the the volume vs. height of the Timber trees. From this graph, there appears to be a positive correlation. 

```python
trees.plot(kind="scatter", x="Height", y= "Volume", alpha=0.8, rot=30)
plt.xlabel("Height (ft)")
plt.ylabel("Volume (cubic ft)")
plt.title("Volume vs. Height of Trees")
```

### 2) Road Data Set

The Road data set contains information of the annual deaths in road accidents for half the US states. 

```python
# Source: Imperial College, London M.Sc. exercise
road = data("road")
```

We first visulaize the number of road deaths that occur within half of the US. From this, we can see that the state with the highest amount of car deaths is California.

```python
road["deaths"].plot(kind="barh", title="Number of Road Deaths in US")
```

```python
road["deaths"].plot(kind="bar", title="Number of Road Deaths in US")
```

The next visualization shows the relationship between the number of deaths, the number of drives, and the daily max temperature. One hypothesis I had before plotting the data was that more car accidents would occur in colder weather due to snow,ice,hail, etc. It however appears that daily max temperature does not have much correlation with death. In fact the number of deaths appears to remain constant for the range of temperature, which is displayed int he top, rightmost graph. Note that the data does not contain information about the temperature measurement system used.

```python
road.columns = ["Deaths", "Drivers", "popden", "rural","Daily Max Temp", "fuel"]
pd.plotting.scatter_matrix(road[["Deaths", "Drivers", "Daily Max Temp"]])
plt.suptitle("Deaths vs. Drivers vs. Daily Max Temp")
```

### 3) Arbuthnot Data Set

This is a PyDataset documentation of Arbuthnot's data on male and female birth ratios in London from 1629-1710. The data for these 81 years showed that in every year there were more male than female christenings. The data set also adds variables of deaths from the plague and total mortality obtained by Campbell and from Creighton (1965).

```python
# Source: Arbuthnot, John (1710). "An argument for Devine Providence, taken from the constant Regularity observ'd in the Births of both Sexes," _Philosophical transactions_, 27, 186-190. Published in 1711.
births = data("Arbuthnot")
```

The plot below shows that there were more Male than Female christenings during each of these 81 years. 

```python
births.plot(x="Year", y=["Males", "Females"])
plt.title("Number of Male vs. Female Christenings")
```

The hexagraph emphasize that the largest number of deaths caused by the plague occurred between 1660-1670. However, most deaths within these 81 years were probably not caused by the plague. We would need more data on the common diseases in this time period to conclude anything further. 

```python
ax = plt.subplot(111)
births.plot(kind="Hexbin", y="Plague", x="Year", gridsize=20, ax=ax)
plt.ylabel("Plaue Deaths")
plt.title("Plague vs. Year")
```

### 4) Birth and Death Rates Data

This data set contains the birth and death rates for 69 countries.

```python
# Source: J. A. Hartigan (1975), _Clustering Algorithms_. John Wiley & Sons, New York.
birth_death = data("birthdeathrates")
```

We first visualize the birth and death rates for different countries with a simple line plot. It seems that the rates follow a similar cyclicle pattern for most countries, except for the last two in the graph. One key pbservation is that birth rates are always higher than death rates in the countries, exluding Finland.

```python
birth_death.plot()
plt.xlabel("Countries")
plt.ylabel("Number of People")
plt.title("Birth/Deaths Rates per Country")
```

The boxplot visualizes the birth and death rates side by side for all countries. It is important to note that it is a good thing that birth rates are typically higher than death rates.  

```python
birth_death.plot(kind="box")
plt.ylabel("Rates")
plt.title("Birth/Death Rates Overall (All countries)")
```

### 5) NCCTG Lung Cancer Data

Survival in patients with advanced lung cancer from the North Central Cancer Treatment Group. Performance scores rate how well the patient can perform usual daily activities.

```python
# Source: Terry Therneau
lung = data("lung")
```

I would expect that as patients go through more treatment, their meal calorie intake and weight loss would both decrease, and thus have some correlation. It seems that there is a cluster of correlation in the scatter plot below, although this plot does not include the proportion of those who did recieve treatment. Something to look into further would to categorize how each different type of treatment contributes to weightloss. 

```python
lung.plot(kind="scatter", y="meal.cal", x="wt.loss")
plt.title("Total Meal Calories vs. Weight Loss")
```

This next histogram shows the compairson Karnofsky performance (bad=0-good=100) rated by the physician vs. the patient. Note, typically the patients do not score between 50-60 and 70-70, regardless of who the rater is.

```python
lung.plot(kind="hist", y=["ph.karno","pat.karno"], alpha = 0.5)
plt.title("Karnofsky performance rated by physician vs. patient")
plt.xlabel("Karnofsky performance (bad=0-good=100)")
```
